```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Part 1 - Implement Basic Data Structures using Numpy, Pandas
# 1. Lists
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]

# 2. Arrays
array1 = np.array(list1)
array2 = np.array(list2)

# 3. Identify their type using type()
list_type = type(list1)
array_type = type(array1)

# 4. Perform mathematical operations
# Multiplication
mult_result = array1 * array2

# Division
div_result = array2 / array1

# Power off
poweroff_result = array1 ** 2

# 5. Combine text with NumPy function
text_output = f"Addition of Two: {array1} + {array2}"

# 6. Implement np.sin(), log(), log2(), np.exp())
sin_result = np.sin(array1)
log_result = np.log(array1)
log2_result = np.log2(array1)
exp_result = np.exp(array1)

# Part 2 - Visualization of Data using Matplotlib, Pyplot Packages
# Generate a Data set for Health Care
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# 3. Scatter Plot
plt.scatter(age, height, label='Height vs Age')
plt.title('Scatter Plot of Height vs Age')
plt.xlabel('Age')
plt.ylabel('Height')
plt.legend()
plt.show()

# 4. Bar Chart
plt.bar(people, weight, color='blue')
plt.title('Bar Chart of Weight for each Person')
plt.xlabel('People')
plt.ylabel('Weight')
plt.show()

# 5. Histogram
plt.hist(age, bins=3, color='green', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
